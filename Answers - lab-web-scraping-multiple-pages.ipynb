# Retrieve an arbitrary Wikipedia page of "Python" and create a list of links on that page

```python
# 2. url: we start with the 'second' page. Show that you can start whenever you want
url = 'https://en.wikipedia.org/wiki/Python'
```

```python
# 3. download html with a get request
response = requests.get(url)
response.status_code # 200 status code means OK!
```

```python
# 4.1. parse html (create the 'soup')
soup = BeautifulSoup(response.content, "html.parser")
```

```python
# 4.2. check that the html code looks like it should
soup
```

```python
#mw-content-text > div.mw-parser-output
pytab =soup.select("#mw-content-text > div.mw-parser-output a")
pytab
```

```python
Links_list = []
for py in pytab:
    link = py.get("href")
    print(link)
    if link is not None:
        if (("/wiki" in link)): 
            Links_list.append(py["href"])
```

# Create a Python list with the top ten FBI's Most Wanted names

```python
# 2. url: we start with the 'second' page. Show that you can start whenever you want
url2 = 'https://www.fbi.gov/wanted/topten'
```

```python
# 3. download html with a get request
response2 = requests.get(url2)
response2.status_code # 200 status code means OK!
```

```python
# 4.1. parse html (create the 'soup')
soup2 = BeautifulSoup(response2.content, "html.parser")
```

```python
#query-results-0f737222c5054a81a120bce207b0446a > ul > li:nth-child(1) > h3 > a
#main-content > div > div:nth-child(2) > div
soup2.select(" h3 > a")[0].get_text()
```

```python
#initialize empty lists
names = []

# define the number of iterations of our for loop
num_iter = len(soup2.select(" h3 > a"))

tClist = soup2.select(" h3 > a")
# iterate through the result set and retrive all the data
for i in range(num_iter):
    names.append(tClist[i].get_text())

print(names)
```
